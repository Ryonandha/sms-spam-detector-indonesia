```python
%%writefile app.py
import streamlit as st
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# --- KODE APLIKASI ANDA ---
st.title("üïµÔ∏è‚Äç‚ôÇÔ∏è Deteksi SMS Penipuan Indonesia")
st.write("Aplikasi ini menggunakan AI untuk menebak apakah SMS itu **Normal**, **Penipuan (Fraud)**, atau **Promo**.")

sms_input = st.text_area("Masukkan isi SMS di sini:", height=150)

if st.button("Cek SMS Ini"):
    if sms_input:
        # Load data & latih model on-the-fly (seperti sebelumnya)
        url = "https://gist.githubusercontent.com/agtbaskara/a1a7017027cc1df9d35cf06e1e5575b7/raw/59870e27ca217d77ac0d8d8dc100551c0dcd14b3/dataset_sms_spam_v2.csv"
        df = pd.read_csv(url)
        df['label'] = df['label'].map({0: 'Normal', 1: 'Penipuan', 2: 'Promo'})
        vec = TfidfVectorizer(max_features=3000)
        X_vec = vec.fit_transform(df['Teks'])
        model = MultinomialNB()
        model.fit(X_vec, df['label'])

        # Prediksi
        input_vec = vec.transform([sms_input])
        prediksi = model.predict(input_vec)[0]
        proba = model.predict_proba(input_vec).max() * 100

        st.subheader("Hasil Analisis:")
        if prediksi == 'Normal':
            st.success(f"‚úÖ SMS ini terlihat AMAN ({prediksi}) - Yakin: {proba:.1f}%")
        elif prediksi == 'Promo':
            st.info(f"üì¢ SMS ini adalah PROMO - Yakin: {proba:.1f}%")
        else:
            st.error(f"üö® HATI-HATI! Ini indikasi PENIPUAN - Yakin: {proba:.1f}%")
    else:
        st.warning("Masukkan teks SMS dulu ya!")
```
